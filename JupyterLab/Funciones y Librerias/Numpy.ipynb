```python
# Tabla de funciones principales de NumPy

#| Categoría | Función | Uso sencillo | Descripción |
#|-----------|---------|--------------|-------------|
#| **Creación de arrays** | `np.array()` | `np.array([1,2,3])` | Crea un arreglo a partir de una lista o tupla. |
#| | `np.zeros()` | `np.zeros((3,3))` | Crea un arreglo lleno de ceros. |
#| | `np.ones()` | `np.ones((2,4))` | Crea un arreglo lleno de unos. |
#| | `np.arange()` | `np.arange(0,10,2)` | Crea un rango de valores con paso definido. |
#| | `np.linspace()` | `np.linspace(0,1,5)` | Genera valores equidistantes entre dos límites. |
#| | `np.eye()` | `np.eye(3)` | Matriz identidad de tamaño \(n \times n\). |
#| **Propiedades** | `np.shape` | `arr.shape` | Devuelve las dimensiones del arreglo. |
#| | `np.size` | `arr.size` | Número total de elementos. |
#| | `np.ndim` | `arr.ndim` | Número de dimensiones. |
#| **Operaciones matemáticas** | `np.sum()` | `np.sum(arr)` | Suma de todos los elementos. |
#| | `np.mean()` | `np.mean(arr)` | Promedio de los elementos. |
#| | `np.std()` | `np.std(arr)` | Desviación estándar. |
#| | `np.min()` | `np.min(arr)` | Valor mínimo. |
#| | `np.max()` | `np.max(arr)` | Valor máximo. |
#| | `np.dot()` | `np.dot(A,B)` | Producto matricial. |
#| | `np.cross()` | `np.cross(A,B)` | Producto cruzado de vectores. |
#| | `np.linalg.inv()` | `np.linalg.inv(A)` | Inversa de una matriz. |
#| | `np.linalg.det()` | `np.linalg.det(A)` | Determinante de una matriz. |
#| **Manipulación de arrays** | `np.reshape()` | `arr.reshape(2,3)` | Cambia la forma del arreglo. |
#| | `np.transpose()` | `arr.T` | Transpone una matriz. |
#| | `np.concatenate()` | `np.concatenate([A,B])` | Une arrays en una dimensión. |
#| | `np.hstack()` | `np.hstack([A,B])` | Une arrays horizontalmente. |
#| | `np.vstack()` | `np.vstack([A,B])` | Une arrays verticalmente. |
#| **Aleatorios** | `np.random.rand()` | `np.random.rand(3,3)` | Matriz con valores aleatorios uniformes [0,1]. |
#| | `np.random.randn()` | `np.random.randn(3,3)` | Valores aleatorios con distribución normal. |
#| | `np.random.randint()` | `np.random.randint(0,10,(3,3))` | Enteros aleatorios en un rango. |
```
