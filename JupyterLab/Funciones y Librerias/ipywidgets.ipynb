```python
import ipywidgets as widgets 
from IPython.display import display
```

#Crear un slider

```python
slider=widgets.IntSlider(value=5, min=0, max=10, step=1, description="valor:2")
display(slider)
```

#crea un botón

```python
boton=widgets.Button (description="ok")
display(boton)
```

```python
def on_button_click(b):
    print(f"Valor del Slider:{slider.value}")
```

```python
boton.on_click(on_button_click)
```

# Tabla de widgets en `ipywidgets`

| Widget | Descripción | Declaración sencilla |
|--------|-------------|----------------------|
| **IntSlider** | Slider para valores enteros | `widgets.IntSlider(value=0, min=0, max=10, step=1, description="Entero:")` |
| **FloatSlider** | Slider para valores decimales | `widgets.FloatSlider(value=0.0, min=0, max=1.0, step=0.1, description="Decimal:")` |
| **IntText** | Caja de texto para números enteros | `widgets.IntText(value=5, description="Número:")` |
| **FloatText** | Caja de texto para números decimales | `widgets.FloatText(value=3.14, description="Pi:")` |
| **Text** | Caja de texto libre | `widgets.Text(value="Hola", description="Texto:")` |
| **Textarea** | Área de texto multilínea | `widgets.Textarea(value="Escribe aquí...", description="Área:")` |
| **Dropdown** | Menú desplegable | `widgets.Dropdown(options=["Opción 1","Opción 2"], value="Opción 1", description="Menú:")` |
| **RadioButtons** | Botones de selección única | `widgets.RadioButtons(options=["A","B","C"], description="Elegir:")` |
| **Checkbox** | Casilla de verificación | `widgets.Checkbox(value=True, description="Activar")` |
| **ToggleButton** | Botón con estado ON/OFF | `widgets.ToggleButton(value=False, description="Switch")` |
| **Button** | Botón simple | `widgets.Button(description="Ejecutar")` |
| **Label** | Etiqueta de texto | `widgets.Label(value="Etiqueta fija")` |
| **Image** | Mostrar imágenes | `widgets.Image(value=open("imagen.png","rb").read(), format="png")` |
| **Play** | Control tipo “play” para animaciones | `widgets.Play(interval=500, value=0, min=0, max=10, step=1)` |

```python
menu=widgets.Dropdown(options=["Opción 1","Opción 2"], value="Opción 1", description="Menú:")
display(menu)
```
