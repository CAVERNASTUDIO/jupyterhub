# Tabla de funciones principales de Matplotlib (`matplotlib.pyplot`)

| Categoría | Función | Uso sencillo | Descripción |
|-----------|---------|--------------|-------------|
| **Creación de gráficos** | `plt.plot()` | `plt.plot(x, y)` | Grafica líneas simples (x vs y). |
| | `plt.scatter()` | `plt.scatter(x, y)` | Grafica puntos dispersos (scatter plot). |
| | `plt.bar()` | `plt.bar(x, y)` | Grafica barras verticales. |
| | `plt.barh()` | `plt.barh(x, y)` | Grafica barras horizontales. |
| | `plt.hist()` | `plt.hist(datos, bins=10)` | Histograma de frecuencias. |
| | `plt.pie()` | `plt.pie(valores)` | Gráfico circular (pie chart). |
| **Configuración de gráficos** | `plt.title()` | `plt.title("Título")` | Añade título al gráfico. |
| | `plt.xlabel()` | `plt.xlabel("Eje X")` | Etiqueta del eje X. |
| | `plt.ylabel()` | `plt.ylabel("Eje Y")` | Etiqueta del eje Y. |
| | `plt.legend()` | `plt.legend()` | Muestra la leyenda de las series. |
| | `plt.grid()` | `plt.grid(True)` | Activa la cuadrícula. |
| | `plt.axis()` | `plt.axis([xmin, xmax, ymin, ymax])` | Define límites de los ejes. |
| **Estilo y formato** | `plt.style.use()` | `plt.style.use("ggplot")` | Aplica un estilo predefinido. |
| | `plt.figure()` | `plt.figure(figsize=(8,5))` | Crea una nueva figura con tamaño definido. |
| | `plt.subplot()` | `plt.subplot(2,1,1)` | Divide la figura en subgráficos. |
| | `plt.subplots()` | `fig, ax = plt.subplots()` | Crea figura y ejes para control avanzado. |
| **Guardado y visualización** | `plt.savefig()` | `plt.savefig("grafico.png", dpi=300)` | Guarda el gráfico en archivo con resolución definida. |
| | `plt.show()` | `plt.show()` | Muestra el gráfico en pantalla (necesario en Jupyter). |


```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
```

```python
plt.figure(figsize=(8,5))
plt.plot(x,y,label="Seno")
plt.title("Funcion Senoidal")
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.legend()
plt.grid(True)
plt.show
```

# Modelos 3D
# Tabla de funciones principales de `Axes3D` (Matplotlib)

| Función / Método | Uso sencillo | Descripción |
|------------------|--------------|-------------|
| **ax.plot3D()** | `ax.plot3D(x, y, z)` | Traza una curva 3D (línea en el espacio). |
| **ax.scatter3D()** | `ax.scatter3D(x, y, z, c=z)` | Grafica puntos dispersos en 3D con colores opcionales. |
| **ax.bar3d()** | `ax.bar3d(x, y, z, dx, dy, dz)` | Grafica barras en 3D (útil para histogramas volumétricos). |
| **ax.contour3D()** | `ax.contour3D(X, Y, Z, 50)` | Dibuja curvas de nivel en 3D. |
| **ax.plot_surface()** | `ax.plot_surface(X, Y, Z, cmap="viridis")` | Grafica una superficie 3D con colores. |
| **ax.plot_wireframe()** | `ax.plot_wireframe(X, Y, Z)` | Grafica una malla (wireframe) de la superficie. |
| **ax.set_xlabel()** | `ax.set_xlabel("X")` | Etiqueta del eje X. |
| **ax.set_ylabel()** | `ax.set_ylabel("Y")` | Etiqueta del eje Y. |
| **ax.set_zlabel()** | `ax.set_zlabel("Z")` | Etiqueta del eje Z. |
| **ax.view_init()** | `ax.view_init(elev=30, azim=45)` | Ajusta el ángulo de vista (elevación y azimut). |
| **ax.set_xlim()** | `ax.set_xlim([xmin, xmax])` | Define límites del eje X. |
| **ax.set_ylim()** | `ax.set_ylim([ymin, ymax])` | Define límites del eje Y. |
| **ax.set_zlim()** | `ax.set_zlim([zmin, zmax])` | Define límites del eje Z. |


```python
# Datos 2
```

```python
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.cos(np.sqrt(X**2 + Y**2))
```

```python
# Crear figura y ejes 3D
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
# Superficie
ax.plot_surface(X, Y, Z, cmap="magma")
# Etiquetas
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.view_init(elev=30, azim=45)

plt.savefig("magma.png",dpi=300)
plt.show()
```
