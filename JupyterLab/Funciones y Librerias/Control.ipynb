# Tabla de funciones principales de `control`

| Función | Uso sencillo | Descripción |
|---------|--------------|-------------|
| **TransferFunction()** | `ctrl.TransferFunction([1],[1,2,1])` | Define una función de transferencia \(G(s)\) a partir de coeficientes de numerador y denominador. |
| **StateSpace()** | `ctrl.StateSpace(A,B,C,D)` | Define un sistema en espacio de estados con matrices \(A,B,C,D\). |
| **step_response()** | `t,y = ctrl.step_response(sys)` | Calcula la respuesta al escalón unitario de un sistema. |
| **impulse_response()** | `t,y = ctrl.impulse_response(sys)` | Calcula la respuesta al impulso de un sistema. |
| **forced_response()** | `t,y,x = ctrl.forced_response(sys, T, U)` | Simula la respuesta de un sistema a una entrada arbitraria \(U(t)\). |
| **bode_plot()** | `ctrl.bode_plot(sys)` | Genera el diagrama de Bode (magnitud y fase vs frecuencia). |
| **nyquist_plot()** | `ctrl.nyquist_plot(sys)` | Genera el diagrama de Nyquist para análisis de estabilidad. |
| **root_locus()** | `ctrl.root_locus(sys)` | Traza el lugar de las raíces del sistema. |
| **pole()** | `ctrl.pole(sys)` | Devuelve los polos del sistema. |
| **zero()** | `ctrl.zero(sys)` | Devuelve los ceros del sistema. |
| **feedback()** | `ctrl.feedback(G,H)` | Crea un sistema en lazo cerrado con función de transferencia \(G/(1+GH)\). |
| **series()** | `ctrl.series(G1,G2)` | Conecta dos sistemas en serie. |
| **parallel()** | `ctrl.parallel(G1,G2)` | Conecta dos sistemas en paralelo. |
| **lsim()** | `t,y,x = ctrl.lsim(sys, U, T)` | Simula la respuesta de un sistema a una entrada definida por el usuario. |
